```python
                            # LISTS
# Creating a list
fruits = ["apple", "banana", "cherry"]
print(fruits)

# Enumerate function
for index,value in enumerate(fruits):
    print(index,value)

# Accessing elements by index
print(fruits[0])  # First element

# Modifying elements
fruits[1] = "orange"  # Change second element
print(fruits)
 
# Check data type
print(type(fruits))

# Get length of list
print(len(fruits))

# Check if element exists in list
print("cherry" in fruits)

# List slicing - get elements from index 1 to 2 (3 is excluded)
print(fruits[1:3])

# Slice from beginning to index 1 (2 is excluded)
print(fruits[:2])

# Slice from index 2 to end
print(fruits[2:])

# Access last element using negative indexing
print(fruits[-1])

# Access second to last element
print(fruits[-2])

# List Methods

# Add element to end of list
fruits.append("mango")
print(fruits)

# Insert element at specific index
fruits.insert(1,"pineapple")
print(fruits)

# Remove first occurrence of specified element (banana was already changed to orange)
fruits.remove("orange")
print(fruits)

# Remove and return last element
fruits.pop()
print(fruits)

# Remove all elements from list
fruits.clear()
print(fruits)

# Sort list in ascending order (Note: list is empty, so no visible change)
fruits.sort()
print(fruits)

# Reverse the order of elements (Note: list is empty, so no visible change)
fruits.reverse()
print(fruits)

# Create a shallow copy of the list
fruits_copy=fruits.copy()
print(fruits_copy)

# Add multiple elements to the end of list
fruits.extend(["apple", "banana", "cherry"])
print(fruits)

# Find index of first occurrence of element
Apple_index=fruits.index("apple")
print(Apple_index)

# Count occurrences of element in list
print(fruits.count("apple"))

# Integer List methods
Numbers=[1,2,3,4,5]

# Add element to end of list
Numbers.append(6)
print(Numbers)

# Insert element at specific index
Numbers.insert(1,7)
print(Numbers)

# Remove first occurrence of specified element
Numbers.remove(1)
print(Numbers)

# Remove and return last element
Numbers.pop()
print(Numbers)

# Remove all elements from list
Numbers.clear()
print(Numbers)

# Sort list in ascending order (Note: list is empty, so no visible change)
Numbers.sort()
print(Numbers)

# Reverse the order of elements (Note: list is empty, so no visible change)
Numbers.reverse()
print(Numbers)

# Create a shallow copy of the list
Numbers_copy = Numbers.copy()
print(Numbers_copy)

# Add multiple elements to the end of list
Numbers.extend([1,2,3,4,5])
print(Numbers)

# Find index of first occurrence of element
index_result = Numbers.index(3)
print(index_result)

# Count occurrences of element in list
count_result = Numbers.count(3)
print(count_result)

# Calculate sum of all elements in the list
print(sum(Numbers))

# Nested List - Lists containing other lists as elements
Nested_List=[1,2,3,4,5]
print(Nested_List)

# Add a list as a single element (creates nested structure)
Nested_List.append([6,7,8,9,10])
print(Nested_List)

# Insert a list at specific index
Nested_List.insert(1,[11,12,13,14,15])
print(Nested_List)

# Remove the nested list element
Nested_List.remove([6,7,8,9,10])
print(Nested_List)
```
