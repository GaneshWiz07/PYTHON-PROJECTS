```python
                            #TUPLES
# TUPLE CHEATSHEET - Complete Guide with Examples

# 1. Creating Tuples
# Empty tuple
empty_tuple = ()
print("Empty tuple:", empty_tuple)

# Tuple with elements
fruits = ("apple", "banana", "cherry")
print("Fruits tuple:", fruits)

# Tuple with one element (note the comma)
single_tuple = ("hello",)
print("Single element tuple:", single_tuple)

# Tuple without parentheses
colors = "red", "green", "blue"
print("Colors tuple:", colors)

# Mixed data types
mixed_tuple = ("apple", 25, True, 3.14)
print("Mixed tuple:", mixed_tuple)

# 2. Accessing Elements
# By index (0-based)
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])

# Slicing
numbers = (1, 2, 3, 4, 5, 6)
print("Slice [1:4]:", numbers[1:4])
print("Slice [:3]:", numbers[:3])
print("Slice [2:]:", numbers[2:])

# 3. Tuple Properties
# Length
print("Length of fruits:", len(fruits))

# Check if item exists
print("Is 'apple' in fruits?", "apple" in fruits)
print("Is 'grape' in fruits?", "grape" in fruits)

# 4. Tuple Methods
# Count occurrences
sample = (1, 2, 3, 2, 2, 4)
print("Count of 2:", sample.count(2))

# Find index of first occurrence
print("Index of 3:", sample.index(3))

# 5. Iterating Through Tuples
# Simple iteration
print("Iterating through fruits:")
for fruit in fruits:
    print(f"- {fruit}")

# With index
print("With index:")
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

# 6. Tuple Unpacking
# Basic unpacking
a, b, c = fruits
print(f"a={a}, b={b}, c={c}")

# With asterisk (rest)
first, *rest = fruits
print(f"First: {first}, Rest: {rest}")

# 7. Tuple Operations
# Concatenation
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined = tuple1 + tuple2
print("Combined tuple:", combined)

# Repetition
repeated = ("hello",) * 3
print("Repeated tuple:", repeated)

# Comparison
print("(1, 2) < (1, 3):", (1, 2) < (1, 3))
print("(1, 2) == (1, 2):", (1, 2) == (1, 2))

# 8. Converting Between Types
# List to tuple
my_list = [1, 2, 3, 4]
list_to_tuple = tuple(my_list)
print("List to tuple:", list_to_tuple)

# Tuple to list
tuple_to_list = list(fruits)
print("Tuple to list:", tuple_to_list)

# String to tuple
string_to_tuple = tuple("hello")
print("String to tuple:", string_to_tuple)

# 9. Nested Tuples
nested = ((1, 2), (3, 4), (5, 6))
print("Nested tuple:", nested)
print("Access nested element:", nested[0][1])

# 10. Common Use Cases
# Coordinates
point = (10, 20)
x, y = point
print(f"Point coordinates: x={x}, y={y}")

# RGB color values
red_color = (255, 0, 0)
print("Red color RGB:", red_color)

# Database records (immutable)
student_record = ("John", 20, "Computer Science", 3.8)
name, age, major, gpa = student_record
print(f"Student: {name}, Age: {age}, Major: {major}, GPA: {gpa}")

# 11. Important Notes
# Tuples are immutable - cannot modify elements
# fruits[1] = "orange"  # This would raise TypeError!
print("Note: Tuples are immutable - cannot change elements after creation")

# However, if tuple contains mutable objects, those can be modified
mutable_content = ([1, 2], [3, 4])
mutable_content[0].append(3)  # This works!
print("Tuple with mutable content:", mutable_content)

```
