```python
# math module for predefined  math functions and constants
import math
class area:
    def triangle(base, height):
        return base * height / 2

    def rectangle(base, height):
        return base * height

    def circle(radius):
        return math.pi * radius**2


```

```python
# This script helps to check system health
import shutil
import psutil
def check_disk_usage(disk):
    du=shutil.disk_usage(disk)
    free=du.free/du.total*100
    return free>20
def check_cpu_usage():
    usage=psutil.cpu_percent(1)
    return usage<75
if not check_disk_usage("/") or not check_cpu_usage():
    print("ERROR!")
else:
    print("Everything is OK!")
```

```python
# Import the csv module for reading and writing CSV files
import csv

# Open and read the CSV file
with open("test.csv","r") as file:
    reader=csv.reader(file)
    # Iterate through each row in the CSV file
    for row in reader:
        # Unpack row values into variables
        name,age,occupation=row
        # Print formatted output for each person
        print(f"Name:{name},Age:{age},Occupation:{occupation}")

# Define new data to be written to the CSV file
new_row = [
    ["Michael", 32, "Architect"],
    ["Lisa", 29, "Product Manager"],
    ["David", 35, "Data Scientist"],
    ["Anna", 26, "UI/UX Designer"]
]

# Open and write to the CSV file
with open("test.csv","w") as file:
    writer=csv.writer(file)
    # Write all rows at once to the CSV file
    writer.writerows(new_row)

# Open and read the CSV file using DictReader
with open("test.csv","r") as file:
    reader=csv.DictReader(file)
    for row in reader:
        name=row["name"]
        age=row["age"]
        occupation=row["occupation"]
        print(f"Name:{name},Age:{age},Occupation:{occupation}")
# Define new data to be written to the CSV file
new_row = [
    [{"name":"Michael","age":32,"occupation":"Architect"},
     {"name":"Lisa","age":29,"occupation":"Product Manager"},
     {"name":"David","age":35,"occupation":"Data Scientist"},]
]
keys=["name","age","occupation"]
with open("test.csv","w") as file:
    writer=csv.DictWriter(file,fieldnames=keys)
    writer.writeheader()
    writer.writerows(new_row)
```

```python
#Regular Expressions and Parsing
# Import regular expressions module
import re
# Sample log entry
log="April 2019: 2019: IP_ADDRESS06 - feest6811[3456] \"POST /incentivize HTTP/1.1\" 302 4622"
# Define regex pattern to extract number between square brackets
regex = r"\[(\d+)\]"
# Search for pattern in log and extract matched group
result = re.search(regex, log)
print(result[1])  # Print the captured number (3456)
# Examples of different regex patterns:
# Match 'H' at start of string
print(re.search(r"^H","Hello"))
# Match 'o' at end of string
print(re.search(r"o$","Hello"))
# Match 'p' followed by any character followed by 'mp'
print(re.search(r"p.mp","pumpkin"))
# Match 'Pump' case-insensitively
print(re.search(r"Pump","pumpkin",re.IGNORECASE))
#Pattern to match all alphabets
print(re.search(r"[a-zA-Z]", "Pumpkin"))
#Pattern to match all alphabets compulsory
print(re.search(r"[a-zA-Z]+","Pumpkin"))
#Pattern to match "Pump" or "pump"
print(re.search(r"[Pp]ump","Pumpkin"))
#Pattern to match "Cat" or "Dog"
print(re.search(r"Cat|Dog", "I like Dogs"))
#Pattern to match "Cat" and "Dog"
print(re.findall(r"Cat|Dog", "I like Dogs and Cats"))

#Repetition Qualifiers
# Match a digit followed by 3 times any character
print(re.search(r"\d{3}","123456"))
# Match a digit followed by 3 to 5 times any character
print(re.search(r"\d{3,5}","123456"))
# Match a digit followed by 3 to 5 times any character
print(re.search(r"\d{3,5}","1234567890"))
# Match 'Py' followed by any characters and ending with 'n'
print(re.search(r"Py.*n","Pygmalion"))
# Match 'Py' followed by any characters and ending with 'n' (greedy match)
print(re.search(r"Py.*n","Python Programming"))
# Match 'Py' followed by lowercase letters and ending with 'n'
print(re.search(r"Py[a-z]*n","Python Programming"))
# Match one or more 'o' followed by one or more 'l'
print(re.search(r"o+l+","goldfish"))
# Match 'ump' with optional preceding 'p'
print(re.search(r"p?ump","umpkin"))
# Match 'ump' with optional preceding 'p'
print(re.search(r"p?ump","pumpkin"))


#Escape Characters
# Match a dot (.) followed by 'com'
print(re.search(r"\.com","welcome"))
# Match a word character (letter, digit, underscore)
print(re.search(r"\w*","This_is_an_example"))
# Match a space character
print(re.search(r"\s*"," This is an example"))
# Match a digit followed by a space followed by a word
print(re.search(r"\d\s\w*","1234567890 abcdef"))

#Capturing Groups
# Search for a pattern of word characters before and after a comma
result=re.search(r"^(\w*),(\w*)$","Lovelace, Ada")
# Print the entire match object
print(result)
# Print all captured groups as a tuple
print(result.groups())
# Print the entire matched string
print(result[0])
# Format and print the second group followed by first group
print("{} {}".format(result[2],result[1]))

# Function to rearrange names from "Last, First" to "First Last" format
def rearrange_name(name):
  # Search for pattern of words separated by comma
  result=re.search(r"^(\w*),(\w*)$",name)
  # If no match found, return original name
  if result is None:
    return name
  # Return rearranged name with first name first
  return "{} {}".format(result[2],result[1])

# Test the rearrange_name function with example names
print(rearrange_name("Lovelace, Ada"))
print(rearrange_name("Ritchie, Dennis"))

# Search for exactly 5 letter sequence
print(re.search(r"[a-zA-Z]{5}","a ghost"))
# Find all 5 letter sequences in the string
print(re.findall(r"[a-zA-Z]{5}","a scary ghost appeared"))
# Find all 5 letter words (using word boundaries)
print(re.findall(r"\b[a-zA-Z]{5}\b","a scary ghost appeared"))
# Find all words between 5 and 10 characters long
print(re.findall(r"\w{5,10}","I really like strawberries"))

# Split string into list using punctuation marks (. ? !) as delimiters

re.split(r"[.?!]","One sentence. Another one? And the last one!")
# Split string and keep the delimiters by using capturing group
re.split(r"([.?!])","One sentence. Another one? And the last one!")

# Replace email addresses with [REDACTED] using regex pattern
# Pattern matches: word chars, dots, %, +, - before @ and word chars, dots, - after @
re.sub(r"[\w.%+-]+@[\w.-]+","[REDACTED]","Received an email for EMAIL.com")

# Rearrange name from "Last, First" to "First Last" format
# Pattern captures words, spaces, dots and hyphens before and after comma
re.sub(r"^([\w .-]*), ([\w .-]*)$",r"\2 \1","Lovelace, Ada")
```
